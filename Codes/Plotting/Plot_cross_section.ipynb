```python
import pygmt
import geopandas as gpd
import pandas as pd
lake= gpd.read_file('/raid2/cg812/Plots/Askja_lake.geojson')
plotted_stations = pygmt.Figure()
topo_data= '/raid5/Iceland/data/IslandsDEM/IslandsDEMv0_10x10m_zmasl_isn2016_PROJECTED.grd'


minlon, maxlon = -17.18, -16.10
minlat, maxlat = 64.75, 65.25

stations1= pd.read_csv('/raid2/cg812/Stations_to_use.csv')

plotted_stations.grdimage(
    grid=topo_data, region= [minlon, maxlon, minlat, maxlat],
    projection='M4i',
    frame=True, cmap= 'gray'
    )

plotted_stations.plot(data= lake, fill= 'lightblue')
#plotted_stations.plot(x=[-16.14, -16.14], y=[65.2, 64.8], pen= '2p')
#plotted_stations.plot(x=[-16.34, -16.34], y=[65.2, 64.8], pen= '2p')
#plotted_stations.plot(x=[-16.54, -16.54], y=[65.2, 64.8], pen= '2p')
#plotted_stations.plot(x=[-16.94, -16.94], y=[65.2, 64.8], pen= '2p')
#plotted_stations.plot(x=[-17.14, -17.14], y=[65.2, 64.8], pen= '2p')
plotted_stations.plot(x=[-16.6, -16.9], y=[65.025, 65.025], pen= '2p')

for _, row in stations1.iterrows():
    plotted_stations.plot(
        x=row.iloc[2],
        y=row.iloc[1], style="t0.3c")

plotted_stations.show()
```
